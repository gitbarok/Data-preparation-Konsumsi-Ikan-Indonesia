```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/clean/data_konsumsi_ikan.csv')
df.head()
```

```python

fig, ax = plt.subplots(figsize=(20,25))
sns.lineplot(data = df, x='tahun', y='jumlah perkapita', hue='provinsi', ax=ax)
plt.show()
```

```python
# [i for i in df.provinsi if i ]
df.loc[df['provinsi'] == 'KALIMANTAN SELATAN'].plot(kind='line', x='tahun', y='jumlah perkapita', xlabel="Tahun", ylabel="total", title='Jumlah provinsi ikan perkapita/tahun provinsi Kalimantan Selatan')
```

```python
fig, ax = plt.subplots(18, 2, figsize=(15,50), sharex=True)
ax = ax.flatten()
i = 0
for provinsi in df['provinsi'].unique():
    loc_prov = df.loc[df['provinsi'] == provinsi]
    sns.lineplot(data = loc_prov, x='tahun', y='jumlah perkapita', ax=ax[i])
    ax[i].set_title(f'Konsumsi ikan perkapita/pertahun provinsi {provinsi}')
    i+=1
plt.tight_layout()
plt.show()
```
